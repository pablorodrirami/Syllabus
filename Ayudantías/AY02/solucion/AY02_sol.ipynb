## Ejercicio 1: DCChrome I
En este ejercicio aprender√°s a usar _stacks_. Los _stacks_ son estructuras de datos secuenciales que funcionan de forma LIFO (Last In, First Out). Esto significa que se comportan como una lista, pero las √∫nicas operaciones que podemos hacer son a√±adir elementos al final, sacar un elemento del final, y ver cu√°l es el elemento que est√° al final. En python las podemos modelar usando las listas normales y su m√©todo `pop`.

Despu√©s de un largo d√≠a de revisar los contenidos de la semana, te das cuenta que el sistema que usa tu navegador para ir hacia atr√°s y hacia adelante se te hace familiar. Te lo imaginas como una pila de p√°ginas, y que al ir hacia atr√°s est√°s sacando el elemento de m√°s arriba de la pila. "¬°¬°Ya lo s√©, es un stack!!" gritas en medio de la sala de estudio del CAI, mientras la gente lentamente se da vuelta para mirarte con caras de confusi√≥n. Huyes lo m√°s r√°pido posible y te encierras en tu pieza a programar tu propio navegador en python, DCChrome.

Para modelar los movimientos hacia atr√°s y adelante, tendr√°s dos stacks, un `back_stack` y un `forward_stack`, donde guardar√°s las p√°ginas hacia atr√°s y las p√°ginas hacia adelante respectivamente. El siguiente dibujo ilustra lo que estamos haciendo ([source](https://lowleveldesign.io/LLD/WebBrowserHistory)):

![stacks](browser-two-stacks.png)

Deber√°s completar el m√©todo `empezar`, modelando adecuadamente lo que pasa con los stacks en cada caso. Imagina que en tu navegador vas del sitio `a` al sitio `b`, y luego del `b` al `c`. Si vuelves hacia atr√°s al sitio `b`, y de ah√≠ vas al sitio `d`, ¬øqu√© pasa con el bot√≥n de ir hacia adelante? ¬øQu√© pasar√≠a con el `forward_stack` en ese caso?


```python
class DCCBrowser:

    def __init__(self):
        self.back_stack = []
        self.forward_stack = []
        self.sitio = "google.com"
    
    def empezar(self):
        while True:
            print(f"{'DCChrome':^62s}")
            print("-" * 62)
            print(f"{self.sitio:^62s}")
            print("-" * 62)
            print()

            sitio_atras = "-"
            sitio_adelante = "-"
            if self.back_stack:
                sitio_atras = self.back_stack[-1]
            if self.forward_stack:
                sitio_adelante = self.forward_stack[-1]
            
            print(f"1. Atr√°s: {sitio_atras}")
            print(f"2. Adelante: {sitio_adelante}")
            print("3. Nuevo sitio")
            print("4. Salir")
            nro = input("Elija una opci√≥n: ")
            
            if nro == "1":
                if not self.back_stack:
                    continue
                atras = self.back_stack.pop()
                self.forward_stack.append(self.sitio)
                self.sitio = atras
            
            elif nro == "2":
                if not self.forward_stack:
                    continue
                adelante = self.forward_stack.pop()
                self.back_stack.append(self.sitio)
                self.sitio = adelante
            
            elif nro == "3":
                self.forward_stack = []
                self.back_stack.append(self.sitio)
                self.sitio = input("Nombre del sitio: ")
            
            elif nro == "4":
                break

            else:
                print("Input incorrecto")

            

browser = DCCBrowser()
browser.empezar()
```

## Ejercicio 2: DCChrome II
En este ejercicio aprender√°s a usar diccionarios. Los diccionarios son estructuras de datos en donde, en vez de asignarle una posici√≥n a cada valor, se le asigna una llave, con la cual se puede acceder al valor. Los valores pueden ser cualquier tipo de dato (incluyendo clases propias), mientras que las llaves solo pueden ser tipos de datos inmutables, por lo que no pueden ser listas, diccionarios, sets ni clases propias.

Programando tu navegador, te das cuenta que tiene un grave error; deja al usuario entrar a cualquier sitio, incluso a sitios prohibidos üò° y con formato inv√°lido. Es por esto que tendr√°s que a√±adirle una verificaci√≥n adicional a la opci√≥n de "Nuevo sitio". Tendr√°s en el atributo `sitios_permitidos` un diccionario que contendr√° nombres de sitios como llave, y el valor asociado ser√° un booleano que indique si el sitio est√° prohibido o no. Cuando el usuario introduzca un nuevo sitio, tendr√°s que tener el cuenta lo siguiente:
- Si el sitio se encuentra en el diccionario y no est√° prohibido, entrar√°s al sitio.
- Si el sitio se encuentra en el diccionario y est√° prohibido, har√°s un `print` de `"Sitio Prohibido"`.
- Si es que no est√° en el diccionario, deber√°s revisar el formato.
    - Si es que el sitio termina en `".com"` es un sitio seguro ‚úÖ, por lo que lo a√±adir√°s al diccionario asociado al valor `True` y entrar√°s.
    - Si es que el sitio termina en `".ru"`, es un sitio inseguro üè¥‚Äç‚ò†Ô∏è, por lo que lo a√±adir√°s al diccionario asociado al valor `False` y har√°s un `print` de `"Sitio Prohibido"`.
    - Si es que el sitio no termina en `".com"` ni en `".ru"`, est√° en un formato inv√°lido por lo que har√°s un `print`de `"Formato Inv√°lido"`.



```python
class DCCBrowser:

    def __init__(self):
        self.back_stack = []
        self.forward_stack = []
        self.sitio = "google.com"
        self.sitios_permitidos = {
            "google.com": True,
            # para no procrastinar
            "youtube.com": False,
            # para no cometer una falta a la √©tica en IIC2233
            "chatgpt.com": False,
            "chat.deepseek.com": False
        }
    
    def empezar(self):
        while True:
            print(f"{'DCChrome':^62s}")
            print("-" * 62)
            print(f"{self.sitio:^62s}")
            print("-" * 62)
            print()

            sitio_atras = "-"
            sitio_adelante = "-"
            if self.back_stack:
                sitio_atras = self.back_stack[-1]
            if self.forward_stack:
                sitio_adelante = self.forward_stack[-1]
            
            print(f"1. Atr√°s: {sitio_atras}")
            print(f"2. Adelante: {sitio_adelante}")
            print("3. Nuevo sitio")
            print("4. Salir")
            nro = input("Elija una opci√≥n: ")

            if nro == "1":
                if not self.back_stack:
                    continue
                atras = self.back_stack.pop()
                self.forward_stack.append(self.sitio)
                self.sitio = atras
            
            elif nro == "2":
                if not self.forward_stack:
                    continue
                adelante = self.forward_stack.pop()
                self.back_stack.append(self.sitio)
                self.sitio = adelante
            
            elif nro == "3":
                sitio = input("Nombre del sitio: ")
                if sitio in self.sitios_permitidos.keys():
                    if self.sitios_permitidos[sitio]:
                        self.forward_stack = []
                        self.back_stack.append(self.sitio)
                        self.sitio = sitio
                    else:
                        print("Sitio Prohibido")
                else:
                    if sitio[-4:] == ".com":
                        self.sitios_permitidos[sitio] = True
                        self.forward_stack = []
                        self.back_stack.append(self.sitio)
                        self.sitio = sitio
                    elif sitio[-3:] == ".ru":
                        self.sitios_permitidos[sitio] = False
                        print("Sitio Prohibido")
                    else:
                        print("Formato Inv√°lido")
            
            elif nro == "4":
                break

            else:
                print("Input incorrecto")

            

browser = DCCBrowser()
browser.empezar()
print(browser.sitios_permitidos)
```

## Ejercicio 3: DCChrome III

En este ejercicio aprenderemos a usar _Named Tuples_. Las _Named Tuples_ son estructuras que nos permiten un funcionamiento similar a una clase, pero sin m√©todos, es decir, solo con atributos. Al igual que las tuplas, son inmutables y permiten acceso por √≠ndice.

Para darle el √∫ltimo toque a tu navegador, crear√°s una _Named Tuple_ que represente a un sitio. Cada sitio tendr√° una url y una direcci√≥n IP. Deber√°s modificar la l√≥gica anterior para que `self.sitio` sea una _Named Tuple_, y que las llaves del diccionario de sitios permitidos tambi√©n lo sean. La IP asociada al sitio est√° dada por la funci√≥n `asignar_ip()`.

```python
from collections import namedtuple
import random

Sitio = namedtuple("sitio", ["url", "ip"])

def asignar_ip(url):
    random.seed(url)
    return f"{random.randint(0, 255)}.{random.randint(0, 255)}.{random.randint(0, 255)}.{random.randint(0, 255)}"

class DCCBrowser:

    def __init__(self):
        youtube = Sitio("youtube.com", asignar_ip("youtube.com"))
        chatgpt = Sitio("chatgpt.com", asignar_ip("chatgpt.com"))
        deepseek = Sitio("chat.deepseek.com", asignar_ip("chat.deepseek.com"))
        google = Sitio("google.com", asignar_ip("google.com"))

        self.back_stack = []
        self.forward_stack = []
        self.sitio = google
        self.sitios_permitidos = {
            google: True,
            # para no procrastinar
            youtube: False,
            # para no cometer una falta a la √©tica en IIC2233
            chatgpt: False,
            deepseek: False
        }
    
    def empezar(self):
        while True:
            print(f"{'DCChrome':^62s}")
            print("-" * 62)
            print(f"{self.sitio.url:^62s}")
            print(f"{self.sitio.ip:^62s}")
            print("-" * 62)
            print()

            sitio_atras = "-"
            sitio_adelante = "-"
            if self.back_stack:
                sitio_atras = self.back_stack[-1]
            if self.forward_stack:
                sitio_adelante = self.forward_stack[-1]
            
            print(f"1. Atr√°s: {sitio_atras if sitio_atras == "-" else sitio_atras.url}")
            print(f"2. Adelante: {sitio_adelante if sitio_adelante == "-" else sitio_adelante.url}")
            print("3. Nuevo sitio")
            print("4. Salir")
            nro = input("Elija una opci√≥n: ")

            if nro == "1":
                if not self.back_stack:
                    continue
                atras = self.back_stack.pop()
                self.forward_stack.append(self.sitio)
                self.sitio = atras
            
            elif nro == "2":
                if not self.forward_stack:
                    continue
                adelante = self.forward_stack.pop()
                self.back_stack.append(self.sitio)
                self.sitio = adelante
            
            elif nro == "3":
                url = input("Nombre del sitio: ")
                ip_sitio = asignar_ip(url)
                sitio = Sitio(url, ip_sitio)
                if sitio in self.sitios_permitidos.keys():
                    if self.sitios_permitidos[sitio]:
                        self.forward_stack = []
                        self.back_stack.append(self.sitio)
                        self.sitio = sitio
                    else:
                        print("Sitio Prohibido")
                else:
                    if sitio.url[-4:] == ".com":
                        self.sitios_permitidos[sitio] = True
                        self.forward_stack = []
                        self.back_stack.append(self.sitio)
                        self.sitio = sitio
                    elif sitio.url[-3:] == ".ru":
                        self.sitios_permitidos[sitio] = False
                        print("Sitio Prohibido")
                    else:
                        print("Formato Inv√°lido")
            
            elif nro == "4":
                break

            else:
                print("Input incorrecto")

            

browser = DCCBrowser()
browser.empezar()
```
